```python
#This is how you import libraries
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this is a random number generator using the random library
r.random()
```

```python
#this is an array of random numbers using numpy
np.random.random((10,10))
```

```python
#this is a scatter plot using matplotlib and numpy
mu, sigma, sampleSize = 1, 0.25, 10000 # mean and standard deviation
plt.scatter(np.random.normal(mu, sigma, sampleSize),np.random.normal(mu, sigma, sampleSize),alpha=.05,c='orange')

plt.show()
```

```python
#this forces the plot to show the origin
mu, sigma, sampleSize = 1, 0.25, 10000 # mean and standard deviation
plt.scatter(np.random.normal(mu, sigma, sampleSize),np.random.normal(mu, sigma, sampleSize),alpha=.05,c='orange')

plt.ylim([0,2])
plt.xlim([0,2])

plt.show()
```
